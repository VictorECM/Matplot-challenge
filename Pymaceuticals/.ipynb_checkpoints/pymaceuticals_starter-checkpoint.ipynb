```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
#forgot to import linregress
# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_clinical_combine = pd.merge(study_results, mouse_metadata, on="Mouse ID")

# Display the data table for preview
mouse_clinical_combine.head()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
summary_stats = df_dup
summary_stats.set_index('Drug Regimen', inplace = True)

mean = summary_stats.groupby('Drug Regimen')["Tumor Volume (mm3)"].mean()
median= summary_stats.groupby('Drug Regimen')["Tumor Volume (mm3)"].median()
variance = summary_stats.groupby('Drug Regimen')["Tumor Volume (mm3)"].var()
std_dev =summary_stats.groupby('Drug Regimen')["Tumor Volume (mm3)"].std()
sem =summary_stats.groupby('Drug Regimen')["Tumor Volume (mm3)"].sem()

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
stats_final ={"Mean":mean,"Median":median,"Variance":variance,"Std Deviation":std_dev,"SEM":sem}
stats_normal =pd.DataFrame(stats_final,index =['Capomulin','Ceftamin','Infubinol','Ketapril','Naftisol','Placebo','Propriva','Ramicane','Stelasyn','Zoniferol'])
stats_normal
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
summary_table =df_dup.groupby("Drug Regimen")["Tumor Volume (mm3)"].agg(['count','mean','median','var','std','sem'])
summary_table
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
mouse = mouse_clinical_combine.loc[mouse_clinical_combine['Mouse ID'] == 's185']

plt.plot(mouse['Timepoint'], mouse['Tumor Volume (mm3)'], marker = 'o')

# Add labels and title to plot
plt.xlabel("Time (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Capomulin Treatment of Mouse s185")

# Display plot
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
capomulin_df =mouse_clinical_combine.loc[mouse_clinical_combine['Drug Regimen']== drug_list[0]]
avg_tumourvol = pd.DataFrame(capomulin_df.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()).reset_index().rename(columns={'Tumor Volume (mm3)': 'avg_tumor_vol'})

avg_tumourvol = pd.merge(capomulin_df, avg_tumourvol, on='Mouse ID')
final_avg_vol_df = avg_tumourvol[['Weight (g)', 'avg_tumor_vol']].drop_duplicates()
final_avg_vol_df.head(5)
                            
plt.scatter(final_avg_vol_df['Weight (g)'],final_avg_vol_df['avg_tumor_vol'],c=final_avg_vol_df['avg_tumor_vol'])
clb = plt.colorbar()
plt.xlabel("Weight (g)")
plt.ylabel("Avg_tumor_vol")
clb.set_label("Tumour Volume")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
x_values = avg_tumourvol['Weight (g)']
y_values = avg_tumourvol['avg_tumor_vol']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20,45))
plt.annotate(f"r^2= {round((rvalue**2),8)}",(21,36))
plt.title('Plot with Avg Wt of mice with Capomulin drug')
plt.xlabel('Weight (g)')
plt.ylabel('Avg_tumor_vol')
plt.show()
```
